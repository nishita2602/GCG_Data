```python
import pandas as pd 
```

```python
sod_register = pd.read_csv('sod_register.csv')
```

```python
# Load the data
register_df = pd.read_csv("sod_register.csv", parse_dates=['event_time'])
session_df = pd.read_csv("sod_session_start.csv", parse_dates=['event_time'])
```

```python
# Step 1: Prepare register data
register_df['register_date'] = register_df['event_time'].dt.date
register_df = register_df[['user_id', 'register_date']]
```

```python
register_df
```

```python
# Step 2: Merge session data with register data
session_df = session_df[['user_id', 'event_time']]
session_df = session_df.merge(register_df, on='user_id', how='left')
session_df
```

```python
# Step 3: Calculate days since registration for each session
session_df['session_day'] = (session_df['event_time'].dt.date - session_df['register_date']).dt.days
session_df
```

```python
# Step 4: Keep only Day 1, 3, 7
session_filtered = session_df[session_df['session_day'].isin([1, 3, 7])]
session_filtered
```

```python
# Step 5: Pivot to get retention flags per user
retention_pivot = session_filtered.pivot_table(
    index=['user_id', 'register_date'],
    columns='session_day',
    values='event_time',
    aggfunc='min'
).reset_index()
retention_pivot
```

```python
retention_pivot.columns.name = None  # remove pivot axis name
retention_pivot
```

```python
retention_pivot.rename(columns={
    1: 'd1_returned',
    3: 'd3_returned',
    7: 'd7_returned'
}, inplace=True)
retention_pivot
```

```python
# Step 6: Mark retained (1 if session exists, 0 otherwise)
retention_pivot['d1_retained'] = retention_pivot['d1_returned'].notnull().astype(int)
retention_pivot['d3_retained'] = retention_pivot['d3_returned'].notnull().astype(int)
retention_pivot['d7_retained'] = retention_pivot['d7_returned'].notnull().astype(int)
retention_pivot
```

```python
# Step 7: Aggregate by register_date
summary_df = retention_pivot.groupby('register_date').agg(
    registered_users=('user_id', 'count'),
    day_1_retained=('d1_retained', 'sum'),
    day_3_retained=('d3_retained', 'sum'),
    day_7_retained=('d7_retained', 'sum')
).reset_index()

```

```python
print(summary_df)
```

```python
# Step 8: Export to CSV for Tableau
summary_df.to_csv("retention_summary_for_tableau.csv", index=False)

print("Exported retention_summary_for_tableau.csv")
```
