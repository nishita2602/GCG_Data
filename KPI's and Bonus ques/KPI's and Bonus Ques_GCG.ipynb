#  2. KPI 1: ARPPU Over Time (Revenue / Paying Users)

```python
# Filter paying users
payers = revenue_df[revenue_df['user_payer_user'] == True]

# Group by date
daily_arppu = payers.groupby('event_date').agg({
    'event_revenue_gross_usd': 'sum',
    'user_id': 'nunique'
}).reset_index()

daily_arppu['arppu'] = daily_arppu['event_revenue_gross_usd'] / daily_arppu['user_id']

print(daily_arppu)
```

```python
avg_arppu = daily_arppu['arppu'].mean()
print(avg_arppu)
```

```python
daily_arppu.to_csv("daily_arppu.csv", index=False)

print("daily_arppu.csv")
```

# KPI's : Installs, LTV, DAU, MAU

```python
import pandas as pd

# Load data with proper datetime parsing
session_df = pd.read_csv("sod_session_start.csv")
register_df = pd.read_csv("sod_register.csv")
revenue_df = pd.read_csv("sod_revenue.csv")
```

```python
# Explicitly convert event_time columns to datetime
session_df['event_time'] = pd.to_datetime(session_df['event_time'], errors='coerce')
register_df['event_time'] = pd.to_datetime(register_df['event_time'], errors='coerce')
revenue_df['event_time'] = pd.to_datetime(revenue_df['event_time'], errors='coerce')
```

```python
# Add date and month columns
session_df['date'] = session_df['event_time'].dt.date
session_df['month'] = session_df['event_time'].dt.to_period('M')

register_df['date'] = register_df['event_time'].dt.date

revenue_df['date'] = revenue_df['event_time'].dt.date
revenue_df['month'] = revenue_df['event_time'].dt.to_period('M')
```

```python
# Daily Active Users
dau = session_df.groupby('date')['user_id'].nunique().reset_index(name='DAU')

# MAU
mau = session_df.groupby('month')['user_id'].nunique().reset_index(name='MAU')

# Installs per day
installs = register_df.groupby('date')['user_id'].nunique().reset_index(name='Installs')

# Revenue per day
daily_revenue = revenue_df.groupby('date')['event_revenue_gross_usd'].sum().reset_index(name='Revenue')

# LTV = Revenue / Installs (per day)
ltv_df = pd.merge(installs, daily_revenue, on='date', how='left')
ltv_df['LTV'] = ltv_df['Revenue'] / ltv_df['Installs']
ltv_df.fillna(0, inplace=True)
```

```python
print(dau)
print(mau)
```

```python
print(installs)
print(ltv_df)
```

```python

# Combine DAU, Installs, Revenue, LTV for export

combined = pd.merge(dau, installs, on='date', how='outer')

combined
```

```python
combined = pd.merge(combined, ltv_df[['date', 'Revenue', 'LTV']], on='date', how='outer')

combined = combined.sort_values(by='date')
```

```python
# Export

combined.to_csv("dau_installs_ltv.csv", index=False)

mau.to_csv("mau.csv", index=False)
```
